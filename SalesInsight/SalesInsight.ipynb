```python
import pandas as pd  # This is our super-tool to read and clean data
```

```python
df = pd.read_csv('/Users/shruti/Documents/SalesInsight/Data/Sample - Superstore.csv', encoding='latin1')
```

```python
df.head()  # Shows the first 5 rows like the first 5 pages of a book
```

```python
df.shape  # Tells you (rows, columns)
```

```python
df.columns
```

```python
df = df.drop_duplicates()
```

```python
df.isnull().sum()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
df['Category'] = df['Category'].str.strip().str.title()
df['Sub-Category'] = df['Sub-Category'].str.strip().str.title()
df['Product Name'] = df['Product Name'].str.strip()
```

```python
df.info()
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# For pretty charts
sns.set(style='whitegrid')
%matplotlib inline
```

```python
df['Month-Year'] = df['Order Date'].dt.to_period('M')
```

```python
#Plot Monthly Sales and Profit Trends
monthly = df.groupby('Month-Year')[['Sales', 'Profit']].sum().reset_index()
monthly['Month-Year'] = monthly['Month-Year'].astype(str)

plt.figure(figsize=(14,6))
sns.lineplot(x='Month-Year', y='Sales', data=monthly, label='Sales')
sns.lineplot(x='Month-Year', y='Profit', data=monthly, label='Profit')
plt.xticks(rotation=45)
plt.title('Sales & Profit Trend Over Time')
plt.xlabel('Month-Year')
plt.ylabel('Amount')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#Top 10 Products by Sales
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,5))
sns.barplot(x=top_products.values, y=top_products.index, hue=top_products.index, palette='viridis', dodge=False, legend=False)
plt.title('Top 10 Products by Sales')
plt.xlabel('Total Sales')
plt.ylabel('Product Name')
plt.show()
```

```python
#Category and Sub-Category Performance
plt.figure(figsize=(10,6))
sns.barplot(x='Category', y='Sales', data=df, estimator=sum, errorbar=None)
plt.title('Sales by Category')
plt.show()

plt.figure(figsize=(14,6))
sns.barplot(x='Sub-Category', y='Profit', data=df, estimator=sum, errorbar=None)
plt.xticks(rotation=45)
plt.title('Profit by Sub-Category')
plt.show()
```

```python
#Region and State Analysis
region_profit = df.groupby('Region')['Profit'].sum().reset_index()

plt.figure(figsize=(8,5))
sns.barplot(x='Region', y='Profit', data=region_profit)
plt.title('Profit by Region')
plt.show()
```

```python
#Discount vs Profit
plt.figure(figsize=(10,6))
sns.scatterplot(x='Discount', y='Profit', data=df)
plt.title('Discount vs Profit')
plt.show()
```

```python
#Top 10 Customers by Sales
top_customers = df.groupby('Customer Name')['Sales'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 5))
sns.barplot(x=top_customers.values, 
            y=top_customers.index, 
            hue=top_customers.index, 
            palette='magma', 
            legend=False)
plt.title('Top 10 Customers by Sales')
plt.xlabel('Total Sales')
plt.ylabel('Customer Name')
plt.show()
```
